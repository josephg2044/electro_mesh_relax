```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mesh(V_b, mask, rho, target=0.001):
    delta = 1.0
    deltas = []
    V_c = V_b.copy()
    while delta > target:
        step = V_c.copy()
        V_c[1:-1, 1:-1] = (step[2:, 1:-1] + step[:-2, 1:-1] +
                         step[1:-1, 2:] + step[1:-1, :-2] + rho[1: -1, 1:-1]) / 4.0
        V_c[mask] = V_b[mask]
        delta = np.max(np.abs(V_c - step))
        deltas.append(delta)
    return V_c, deltas
```

```python
def make_grid(V):
    M = V.shape[0] - 1
    x = np.linspace(0, M, M + 1)
    y = np.linspace(0, M, M + 1)
    X, Y = np.meshgrid(x, y)
    return X, Y, y

def heatmap(V):
    fig, ax = plt.subplots()
    ax.imshow(V, cmap="inferno", origin="lower")

def contour(V, X, Y):
    fig, ax = plt.subplots()
    CS = ax.contour(X, Y, V, 6, cmap="GnBu", levels=21, linewidths=0.75)
    ax.clabel(CS, fontsize=9)
    ax.imshow(V, cmap="inferno", origin="lower")

def vol(V, X, Y):
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    surf = ax.plot_surface(X, Y, V, cmap="inferno", linewidth=0, antialiased=False)

def field(V, X, Y, y):
    fig, ax = plt.subplots()
    Ey, Ex = np.gradient(-V)
    ax.streamplot(X, Y, Ex, Ey, color="black", linewidth=0.5)
    ax.contour(X, Y, V, colors = "black", linewidths=0.5)
    ax.axis("square")


def plot_error(errors):
    fig, ax = plt.subplots()
    ax.plot(errors, marker="", linestyle="-", color="red")
    ax.set_yscale("log")
    ax.set_xlabel('iterations')
    ax.set_ylabel('delta value')

def make_plots(V, errors):
    X, Y, y = make_grid(V)
    
    contour(V, X, Y)
    vol(V, X, Y)
    field(V, X, Y, y)
    plot_error(errors)
```

```python
M = 100

V_b = np.zeros([M + 1, M + 1], float)
V_b[-1,:] = 10
mask_b = np.zeros((M+1, M+1), dtype=bool)
mask_b[0, :] = mask_b[-1, :] = mask_b[:, 0] = mask_b[:, -1] = True
rho_b = np.zeros([M + 1, M + 1], float)

V_d = np.zeros([M + 1, M + 1], float)
mask_d = np.zeros((M+1, M+1), dtype=bool)
mask_d[0, :] = mask_d[-1, :] = mask_d[:, 0] = mask_d[:, -1] = True
rho_d = np.zeros([M + 1, M + 1], float)
rho_d[20:40, 20:40] = -2
rho_d[60:80, 60:80] = 2

V_q = np.zeros([M + 1, M + 1], float)
r, c = np.indices(V_q.shape)
V_q[(r - 50)**2 + (c - 30)**2 <= 7.5**2] = 10
V_q[(r - 50)**2 + (c - 70)**2 <= 7.5**2] = 10
V_q[(r - 30)**2 + (c - 50)**2 <= 7.5**2] = -10
V_q[(r - 70)**2 + (c - 50)**2 <= 7.5**2] = -10
mask_q = np.zeros((M+1, M+1), dtype=bool)
mask_q[0, :] = mask_q[-1, :] = mask_q[:, 0] = mask_q[:, -1] = True
mask_q[V_q != 0] = True
rho_q = np.zeros([M + 1, M + 1], float)

V_t = np.zeros([M + 1, M + 1], float)
r, c = np.indices(V_q.shape)
V_t[-(c - 35) > np.abs(r - 50)] = 5
V_t[(c < 15)] = 0
mask_t = np.zeros((M+1, M+1), dtype=bool)
mask_t[0, :] = mask_t[-1, :] = mask_t[:, 0] = mask_t[:, -1] = True
mask_t[V_t != 0] = True
rho_t = np.zeros([M + 1, M + 1], float)
rho_t[25:35, 55:85] = -48/25
rho_t[65:75, 55:85] = -48/25
rho_t[35:65, 75:85] = -48/25
```

```python
make_plots(*mesh(V_b, mask_b, rho_b))
make_plots(*mesh(V_d, mask_d, rho_d))
make_plots(*mesh(V_q, mask_q, rho_q))
make_plots(*mesh(V_t, mask_t, rho_t))
```
